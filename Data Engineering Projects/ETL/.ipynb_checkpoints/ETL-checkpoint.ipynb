## Transform phase

1. Cleaning data
2. User_defined rules

### Data Cleaning

```python
# Customer Data

#nulls
customers.isna().sum()
```

```python
#duplicates
print(len(customers))
customers.drop_duplicates(inplace=True)
print(len(customers))
```

```python
# products Data

#nulls
products_men.isna().sum()
```

```python
#get important columns only
products_men = products_men[["url","name","offers/0/price","sku"]]
```

```python
#rename columns
products_men.rename(columns={"offers/0/price":"price","sku":"product_id"},inplace=True)
```

```python
#nulls
products_women.isna().sum()
```

```python
#get important columns only
products_women = products_women[["url","name","offers/0/price","sku"]]
```

```python
#rename columns
products_women.rename(columns={"offers/0/price":"price","sku":"product_id"},inplace=True)
```

```python
#union the two dataframes
products = pd.concat([products_men, products_women], ignore_index=True)
```

```python
# Sales Data

#nulls
sales.isna().sum()
```

```python
#duplicates
print(len(sales))
sales.drop_duplicates(inplace=True)
print(len(sales))
```

```python
#making sure consistent naming convention

customers.columns = customers.columns.str.lower()
products.columns = products.columns.str.lower()
sales.columns = sales.columns.str.lower()
```

### User_defined rules

```python
# Apply User-Defined Rules

#identify high value customer
customers["high_value_customer"] = customers["spending_score"] > 50

#identify high end products
products["high_value_product"] = products["price"] > 90

#identify high value sales
sales["high_value_sale"] = sales["quantity"] > 100
```

## Load Phase

```python
# MySQL connection
conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="your_password",
    database="sales_db"
)
cursor = conn.cursor()

# Create table
cursor.execute("""
CREATE TABLE IF NOT EXISTS sales_data (
    TransactionID INT,
    CustomerID INT,
    ProductID INT,
    Quantity INT,
    SaleDate DATE,
    TotalSaleAmount FLOAT,
    HighValueSale BOOLEAN,
    Region VARCHAR(255),
    InvalidRegion BOOLEAN
)
""")

# Insert data
for _, row in sales.iterrows():
    cursor.execute("""
    INSERT INTO sales_data (TransactionID, CustomerID, ProductID, Quantity, SaleDate, TotalSaleAmount, HighValueSale, Region, InvalidRegion)
    VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s)
    """, tuple(row))

#close cursor and connection
conn.commit()
cursor.close()
conn.close()
```
